# Single group single column 
 # Count of passengers who survived from each class 


```python
Pclass_survived= titanic.groupby(['Pclass'])['Survived']
Pclass_survived.sum()
```

# Multiple column
# Total amount paid by each class who survived

```python
pclass_survived_fare=titanic.groupby(['Pclass'])['Survived','Fare']
pclass_survived_fare.sum()
```

# Multiple groups 
# Count of females and males who survived in each class 

```python
pclass_survived_sex=titanic.groupby(['Pclass','Sex'])['Survived']
pclass_survived_sex.sum()
```

```python
pclass_survived_sex=titanic.groupby(['Pclass','Sex'], as_index=False)['Survived']
pclass_survived_sex.sum()
```

# Multiple functions 
# Mean and sum of pclass & survived 


```python
pclass_survived_mean_sum=titanic.groupby(['Pclass'])['Survived'].agg(['mean','sum'])
pclass_survived_mean_sum
```
