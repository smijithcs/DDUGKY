```python
netflix=pd.read_csv("netflix_titles_2021.csv")
```

```python
netflix
```

```python
#assigning new column
netflix.assign(after_2020=netflix['release_year']>2020)
```

```python
#assigning new column method 2
netflix['after_2019']=netflix['release_year']>2019
netflix
```

```python
# query movies released in 2019 only 
netflix.query('release_year==2019 & type=="Movie"')
```

```python
#checking for null values
netflix.isnull().sum()
```

```python
#removing rows with null values with respect to cloumns
netflix=netflix.dropna(subset=["director","cast"])
netflix
```

```python
netflix.drop(columns=['country','cast'])
```

```python
pivot=pd.pivot_table(columns='rating',values='release_year',data=netflix,aggfunc="median")
pivot
```
