## Bar plot
### To create the below shown bar chart we have to create the frequency table. Frequency table can be made using the value_count method

```python
data.Manufacturer.value_counts().plot.bar(figsize=(6,2), cmap='Accent');
```

```python
data.sort_values(by='Price_in_thousands')[:20].plot.bar('Manufacturer','Price_in_thousands',figsize=(6,3),cmap='Accent')
```

```python
data.sort_values(by='Price_in_thousands')[:20].plot.scatter('Latest_Launch','Price_in_thousands',figsize=(12,5),cmap='brg');
```
